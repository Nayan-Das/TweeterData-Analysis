```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
from nltk.stem import PorterStemmer
import string
```

```python
with open('training_hindi_to_english_16932.txt', 'r') as f:
    l = list()
    for i in iter(f):
        l.append(i.strip('\n').split('  ->  '))
```

```python
import unicodedata
new_l = list()
for i in l:
    strings = unicodedata.normalize("NFKD", i[1])
    while True:
        loc1 = strings.find('http')
        if loc1 == -1: break
        sub1 = strings[: loc1]
        sub2 = strings[loc1 : ]
        loc2 = sub2.find(' ')
        if loc2 == -1: strings = sub1
        else: strings = sub1 + strings[loc1 + loc2 : ]
    while True:
        loc1 = strings.find('@')
        if loc1 == -1: break
        sub1 = strings[: loc1]
        sub2 = strings[loc1 : ]
        loc2 = sub2.find(' ')
        if loc2 == -1: strings = sub1
        else: strings = sub1 + strings[loc1 + loc2 : ]
    while True:
        loc1 = strings.find('#')
        if loc1 == -1: break
        sub1 = strings[: loc1]
        sub2 = strings[loc1 : ]
        loc2 = sub2.find(' ')
        if loc2 == -1: strings = sub1
        else: strings = sub1 + strings[loc1 + loc2 : ]
    #print(i)
    #print([strings.strip()], end="\n\n")
    new_l.append([i[0], strings.strip().lower()])
```

```python
punct = set(string.punctuation)
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
match = string.ascii_lowercase
```

```python
l_store = list()
for i in new_l:
    word_tokens = word_tokenize(i[1])
    #print(word_tokens)
    filtered_sentence = [w for w in word_tokens if w not in stop_words]
    #print(filtered_sentence)
    filtered_sentence = [w for w in filtered_sentence if not (len(w) == 1 and w in punct)]
    #print(filtered_sentence)
    filtered_sentence = [w for w in filtered_sentence if any([c in match for c in w]) == True]
    #print(filtered_sentence)
    filtered_sentence = ' '.join([stemmer.stem(w) for w in filtered_sentence])
    #print(filtered_sentence)
    l_store.append([filtered_sentence])
```

```python
comment_words = ' '
stopwords = set(STOPWORDS) 
```

```python
# iterate through the csv file 
for val in l_store: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
    tokens = [w for w in tokens if tokens != 'rt']
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
          
    for words in tokens: 
        comment_words = comment_words + words + ' '
  
  
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 
```

```python
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```
