```python
with open('training_hindi_to_english_16932.txt', 'r') as f:
    l = list()
    for i in iter(f):
        l.append(i.strip('\n').split('  ->  '))
```

```python
## removing web - links, # - tag, @ - reference
```

```python
import unicodedata
new_l = list()
for i in l:
    strings = unicodedata.normalize("NFKD", i[1])
    while True:
        loc1 = strings.find('http')
        if loc1 == -1: break
        sub1 = strings[: loc1]
        sub2 = strings[loc1 : ]
        loc2 = sub2.find(' ')
        if loc2 == -1: strings = sub1
        else: strings = sub1 + strings[loc1 + loc2 : ]
    while True:
        loc1 = strings.find('@')
        if loc1 == -1: break
        sub1 = strings[: loc1]
        sub2 = strings[loc1 : ]
        loc2 = sub2.find(' ')
        if loc2 == -1: strings = sub1
        else: strings = sub1 + strings[loc1 + loc2 : ]
    while True:
        loc1 = strings.find('#')
        if loc1 == -1: break
        sub1 = strings[: loc1]
        sub2 = strings[loc1 : ]
        loc2 = sub2.find(' ')
        if loc2 == -1: strings = sub1
        else: strings = sub1 + strings[loc1 + loc2 : ]
    print(i)
    print([strings.strip()], end="\n\n")
    new_l.append([i[0], strings.strip().lower()])
```

```python
#stop word removal, removal of punctuations, stemming using nltk
#do not apply lemmatization
```

```python
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
from nltk.stem import PorterStemmer
import string
```

```python
punct = set(string.punctuation)
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
match = string.ascii_lowercase
```

```python
l_store = list()
```

```python
for i in new_l:
    word_tokens = word_tokenize(i[1])
    #print(word_tokens)
    filtered_sentence = [w for w in word_tokens if w not in stop_words]
    #print(filtered_sentence)
    filtered_sentence = [w for w in filtered_sentence if not (len(w) == 1 and w in punct)]
    #print(filtered_sentence)
    filtered_sentence = [w for w in filtered_sentence if any([c in match for c in w]) == True]
    #print(filtered_sentence)
    filtered_sentence = ' '.join([stemmer.stem(w) for w in filtered_sentence])
    #print(filtered_sentence)
    l_store.append([i[0], filtered_sentence])
```

```python
with open('processed_training_data.txt', 'w')  as f:
    for i in l_store:
        f.writelines(str(i[0]) + '  ->  ' + str(i[1]) + '\n')
```
